```python
from sklearn.datasets import load_iris, load_wine
from sklearn.cluster import KMeans, DBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

```

```python
# Load the iris dataset
iris = load_iris()

# Load the wine dataset
wine = load_wine()

# Extract the features
df = pd.DataFrame(iris.data, columns=iris.feature_names)
data = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)
Y = scaler.fit_transform(data)
```

```python
# Initialize the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X)
```

```python

# Add the cluster labels to the dataset
labels = pd.DataFrame(kmeans.labels_, columns=['cluster'])

df1 = df
```

```python
df1
```

```python
df1 = pd.concat([df1, labels], axis=1)

# Save the dataset
#df1.to_csv('iris_kmeans.csv', index=False)
df1
```

```python
# Initialize the AgglomerativeClustering model
agglo = AgglomerativeClustering(n_clusters = 3)

# Fit the model to the data
agglo.fit(X)
```

```python
# Add the cluster labels to the dataset
labels = pd.DataFrame(agglo.labels_, columns=['cluster'])
df2 = df
```

```python
# split datasets into training and testing sets
X_train1, X_test1, y_train1, y_test1 = train_test_split(df1.drop('cluster', axis=1), df1['cluster'], test_size=0.3, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(df2.drop('cluster', axis=1), df2['cluster'], test_size=0.3, random_state=42)
X_train3, X_test3, y_train3, y_test3 = train_test_split(df3.drop('cluster', axis=1), df3['cluster'], test_size=0.3, random_state=42)
X_train4, X_test4, y_train4, y_test4 = train_test_split(df4.drop('cluster', axis=1), df4['cluster'], test_size=0.3, random_state=42)

# apply supervised methods on dataset 1
lr1 = LogisticRegression()
dt1 = DecisionTreeClassifier()
rf1 = RandomForestClassifier()
svm1 = SVC()
models1 = [lr1, dt1, rf1, svm1]
for model in models1:
    model.fit(X_train1, y_train1)
    y_pred1 = model.predict(X_test1)
    accuracy1 = accuracy_score(y_test1, y_pred1)
    print(f"Accuracy score of {type(model).__name__} on dataset 1: {accuracy1}")
    
print('\n')

# apply supervised methods on dataset 2
lr2 = LogisticRegression()
dt2 = DecisionTreeClassifier()
rf2 = RandomForestClassifier()
svm2 = SVC()
models2 = [lr2, dt2, rf2, svm2]
for model in models2:
    model.fit(X_train2, y_train2)
    y_pred2 = model.predict(X_test2)
    accuracy2 = accuracy_score(y_test2, y_pred2)
    print(f"Accuracy score of {type(model).__name__} on dataset 2: {accuracy2}")
    
print('\n')
    
# apply supervised methods on dataset 3
lr3 = LogisticRegression()
dt3 = DecisionTreeClassifier()
rf3 = RandomForestClassifier()
svm3 = SVC()
models3 = [lr3, dt3, rf3, svm3]
for model in models3:
    model.fit(X_train3, y_train3)
    y_pred3 = model.predict(X_test3)
    accuracy3 = accuracy_score(y_test3, y_pred3)
    print(f"Accuracy score of {type(model).__name__} on dataset 3: {accuracy3}")
    
print('\n')
    
# apply supervised methods on dataset 4
lr4 = LogisticRegression()
dt4 = DecisionTreeClassifier()
rf4 = RandomForestClassifier()
svm4 = SVC()
models4 = [lr4, dt4, rf4, svm4]
for model in models4:
    model.fit(X_train4, y_train4)
    y_pred4 = model.predict(X_test4)
    accuracy4 = accuracy_score(y_test4, y_pred4)
    print(f"Accuracy score of {type(model).__name__} on dataset 4: {accuracy4}")
```
